```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import make_blobs
x, _ =make_blobs(n_samples=500,n_features=2,centers=3,random_state=20)
```

```python
x[:]
```

```python
X=x[:,0]
X
```

```python
y=x[:,1]
y
```

```python
df=pd.DataFrame(X,y)
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.scatter(X,y)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
model=DBSCAN(eps=1,min_samples=5).fit(x)
```

```python
cluster=model.labels_
```

```python
len(set(cluster))
```

```python
y_predicted=model.fit_predict(x)
```

```python
df_cluster=pd.DataFrame(y_predicted)
```

```python
df['cluster']=y_predicted
```

```python
import seaborn as sns
plt.figure(figsize=(8,8))
sns.scatterplot(X,y,hue=cluster,palette='deep')
```
